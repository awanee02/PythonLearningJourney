Exercise 2: List Comprehension
Given this list:
Pythonnumbers = [10, 15, 20, 25, 30, 35, 40, 45, 50]

Create new lists using comprehension:

All numbers greater than 25, All numbers multiplied by 2, All even numbers

```python
Pythonnumbers = [10, 15, 20, 25, 30, 35, 40, 45, 50]

#Number greater than 25
number_greater_than_25 = [x for x in Pythonnumbers if x > 25]

#All numbers multiplied by 2
multiply_by_2 = [x * 2 for x in Pythonnumbers]

#All even numbers
even_numbers = [x for x in Pythonnumbers if x%2 == 0]

print(number_greater_than_25)
print(multiply_by_2)
print(even_numbers)
```

Today's Mini-Project: Student Grade Analyzer

Create a program that: Has a tuple of subjects: ("Math", "Science", "English", "History", "Computer")

Asks the user to enter marks for each subject (store in a dictionary: subject → marks)

Then using list/dictionary comprehensions, calculate and show:
Average marks

Highest scoring subject

Lowest scoring subject

List of subjects where marks > 80 (using comprehension)

Make it nice with good output formatting.

```python
subjects = ("Maths", "Science", "English", "History", "Computer")
marks = {}

print("Enter marks for each subject: \n")

#Taking user input as marks for each individual subject
for subject in subjects:
  score = int(input(f"Enter the marks for {subject}:  "))
  marks[subject] = score

total = sum(marks.values())
average = total / len(marks)

print(average)

highest_subject = max(marks, key=marks.get)
highest_score = marks[highest_subject]
print(f"Highest scoring subject is {highest_subject} with marks {highest_score}")

lowest_subject = min(marks, key=marks.get)
lowest_score = marks[lowest_subject]
print(f"Lowest score subject is {lowest_subject} with marks {lowest_score}")

marks_greater_than_80 = [subject for subject in subjects if score > 80]
print(marks_greater_than_80)
```
