Exercise 3: Dynamic Name Collector

Create an empty list names = []

Use a while loop: keep asking user to enter names until they type "stop".

Then print all entered names.

```python
names = []
print('Enter names one by one. Type "stop" to end the list')

while True:
  name = input(print("Please enter a name: "))

  if name.lower() == "stop":
    break
  names.append(name)

print("Here are all the names you entered: \n")
for name in names:
  print(name)
```

Today's Mini-Project: Simple To-Do List Manager

```python
todo = []   #empty list

while True:
  print("\n------My To-Do List------")
  print("1. Learn a new concept")
  print("2. Practice the older concepts")
  print("3. Remove a task")
  print("4. Exit")

  choice = input("Choose an option in (1-4): ")

  if choice == "1":
    task = input("Enter a new task: ")
    todo.append(task)
    print("Task added!")

  elif choice == "2":
    if not todo:
      print("Your list is empty!")
    else:
      for i, task in enumerate(todo):
        print(f"{i+1}. {task}")

  elif choice == "3":
    if not todo:
      print("Your list is empty!")
    else:
      num = int(input("Enter a task number to remove: "))
      if 1 <= num <= len(todo):
        removed = todo.pop(num-1)
        print(f"Removed: {removed}")
      else:
        print("Invalid number")

  elif choice == "4":
    print("Goodbye!")
    break
```
