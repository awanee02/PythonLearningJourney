File Handling

Exercise 1: Basic File Write

Write a function that creates a file called my_info.txt and writes the following:
Name: Arnav

Age: 25

City: Pune

Learning: Python

```python
def create_my_info():
  with open("my_info.txt", "w")as file:
    file.write("Name: Arnav\n")
    file.write("Age: 22\n")
    file.write("City: Pune\n")
    file.write("Learning Python: Yes\n")
  print("my_info.txt created successfully!")
create_my_info()     #Call the function
```

Exercise 2: Read and Display

Write code to read the file my_info.txt and print its content nicely.

```python
print("------Reading my_info.txt-----\n")

with open("my_info.txt", "r")as file:
  content = file.read()
  print(content)
```

Today's Mini Project: Todo List with File

Create a simple program that:

Allows user to add tasks (saved to todo.txt)
Shows all saved tasks

```python
while True:
  print("------My ToDo List-------")
  print("1. Add a new task")
  print("2. View all tasks")
  print("3. Clear all tasks")
  print("4. Exit")

  choice = input("Choose an option from 1-4: ")

  if choice == '1':
    task = input("Add a new task: ")
    if task:
      with open("todo.txt", "a")as file:
        file.write(task + "\n")
      print("Task added successfully!")

  elif choice == '2':
    try:
      with open("todo.txt", "r")as file:
        tasks = file.readlines()

      if not tasks:
        print("Your todo list is empty.")
      else:
        print("Your todo tasks: ")
        for i, task in enumerate(task,1):
          print(f"{i}. {task.strip()}")
    except FileNotFoundError:
      print("No tasks yet, file doesn't exist.")

  elif choice == '3':
    confirm = input("Are you sure you want to clear all tasks? (yes/no): ").lower()
    if confirm == "yes":
      with open("todo.txt", "w")as file:
        file.write("")
    print("All tasks cleared.")

  elif choice == '4':
    print("Exit.")
    break
```
