Exercise 2: Calculator Functions

Create these 4 functions:

add(a, b)

subtract(a, b)

multiply(a, b)

divide(a, b)

Test them with some numbers.

```python
def add(a,b):
  result = a + b
  return result

def subtract(a,b):
  result = a - b
  return result

def multiply(a,b):
  result = a * b
  return result

def divide(a,b):
  result = a / b
  return result

sum = add(10,20)
difference = subtract(35,25)
product = multiply(20,30)
quotient = divide(20,5)

print("sum = ", sum)
print("difference = ", difference)
print("product = ", product)
print("quotient = ", quotient)
```

Today's Mini-Project: Simple Calculator with Menu

Create a program with these functions:

add(a, b)

subtract(a, b)

multiply(a, b)

divide(a, b)

Then build a menu-driven calculator using while True:

Menu options:

Addition

Subtraction

Multiplication

Division

Exit

Ask user for choice → ask for two numbers → call the correct function → show result.

```python
def add(a,b):
  return a + b

def subtract(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  if b == 0:
    return "Cannot divide by zero!"
  return a / b

while True:
  print("Simple calculator")
  print("1. Addition")
  print("2. Subtraction")
  print("3. Multiplication")
  print("4. Division")
  print("5. Exit")

  choice = input("Choose an option from (1-5): ")

  if choice == "5":
    print("Goodbye!")
    break

  #Ask for two numbers
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))

  #Call function based on choice
  if choice == "1":
    result = add(num1, num2)
    print(f"{num1} + {num2} = {result}")

  elif choice == "2":
    result = subtract(num1, num2)
    print(f"{num1} - {num2} = {result}")

  elif choice == "3":
    result = multiply(num1, num2)
    print(f"{num1} * {num2} = {result}")

  elif choice == "4":
    result = divide(num1, num2)
    print(f"{num1} / {num2} = {result}")

  else:
    print("Invalide choice, please select an option again.")
```
