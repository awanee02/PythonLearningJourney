MINI PRJECT 1: PASSWORD GENERATOR

```python
import random
import string

def get_character_pool(use_upper, use_digits, use_symbol):
  """
  Build a pool of characters to pick from,
  based on what the user wants in their password.

  We include lowercase characters as base,
  then add any other forms of character the user wants.
  """

  #Starting with lowercase characters as the base
  pool = list(string.ascii_lowercase)

  #List for keeping a track of all the characters added from all types
  guaranteed = []

  if use_upper:
    pool += list(string.ascii_uppercase)
    guaranteed.append(random.choice(string.ascii_uppercase))

  if use_digits:
    pool += list(string.digits)
    guaranteed.append(random.choice(string.digits))

  if use_symbol:
    pool += list(string.punctuation)
    guaranteed.append(random.choice(string.punctuation))

  return pool, guaranteed

def generate_password(length, use_upper, use_digits, use_symbol):
  """
  Generate a secure password.
  STRATEGY:
  -Start with one guaranteed character from each character type.
  - Fill the remaining slots with random picks from the pool.
  - Shuffle so that the characters are always not at the same position.
  """

  pool, guaranteed = get_character_pool(use_upper, use_digits, use_symbol)

  #Calculating the remaining length
  remaining_length = length - len(guaranteed)

  #Picking random characters
  pick_random = [random.choice(pool) for _ in range(remaining_length)]

  #Combining all characters into one list
  combine_all = guaranteed + pick_random

  #Shuffling the characters
  random.shuffle(combine_all)

  #Joining the list into a single string
  password = "".join(combine_all)
  return password

def get_yes_no(prompt):
  """
  A small helper function to get yes/no from the user.
  Keep asking until they type y/n.
  """

  while True:
    answer = input(prompt).strip().lower()
    if answer in ("y","n"):
      return answer == "y"
    print("Please enter y or n. ")

def get_valid_length():
  """
  Ask for password length and validate it.
  Must be a number and must be atleast 6.
  """

  while True:
    try:
      length = int(input("Enter password length(minimum 6): "))
      if length < 6:
        print("Password must be atleast 6 characters long. Please try again.")
      else:
        return length
    except ValueError:
      print("Please enter a valid number.")

def main():
  print("-------------SECURE PASSWORD GENERATOR-----------------")

  while True:
    print()

    #STEP 1: Get password length
    length = get_valid_length()

    #STEP 2: Get character type preference
    print()
    use_upper = get_yes_no("Include uppercase characters? (y/n): ")
    use_digits = get_yes_no("Include digits? (y/n): ")
    use_symbol = get_yes_no("Include special characters? (y/n): ")

    #STEP 3: Ask how many passwords to generate
    print()
    while True:
      try:
        count = int(input("How many passwords to generate? (1-10): "))
        if 1 <= count <= 10:
          break
        print("Please enter a number between 1 to 10. ")
      except ValueError:
        print("Please enter a valid number. ")

    #STEP 4: Generate and display passwords
    print()
    print(f" Generated password{'s' if count > 1 else ''}: ")
    for i in range(count):
      pwd = generate_password(length, use_upper, use_digits, use_symbol)
      print(f" {i+1}. {pwd}")

    #STEP 5: Ask if the user wants to generate more passwords
    print()
    again = get_yes_no("Generate more passwords? (y/n): ")
    if not again:
      print()
      print("Stay secure, goodbye!")
      break

if __name__ == "__main__":
    main()
```
