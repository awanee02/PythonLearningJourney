Today's Mini-Project: "Personalized Greeting + Age Calculator"
Write a program that does the following:

Asks the user 4 things:
Their name

Their age (as number)

Their city

Current year

Calculates birth year (current_year - age)

Prints a nice introduction using f-strings only (no old print("text", variable) style)

```python
name = str(input('What is your name? '))
age = int(input('How old are you? '))
city = str(input('Which city do you live in? '))
current_year = int(input('Current year: '))
birth_year = current_year - age
print(f"Welcome to my first interactive program!")
print(f"Hello {name}! You are {age} years old living in {city} and you are born in {birth_year}.")
```
