Conditional Statements

```python
age = int(input('What is your age? '))
if(age >= 18):
  print('You are an adult')
elif(age >= 13):
  print('You are a teenager')
else:
  print('You are a child')
```

Simple If

```python
age = int(input('Tell me your age: '))
if(age >= 18):
  print('You are eligible to vote')
else:
  print('You are not eligible to vote')
```

Multiple Conditions combined with And/or

```python
marks = int(input('Score of the student in the range of 0-100: '))
if(marks >= 90):
  print('The student has got A grade')
elif(marks >=80 & marks < 90):
  print('Student has got B grade')
elif(marks >= 65 & marks < 80):
  print('Student has got C grade')
elif(marks >= 40 & marks < 65):
  print('Student has got D grade')
else:
  print('Student has not cleared the exam')
```

Exercise 1: Number Checker

Ask the user for a number.

Tell them if it is Positive, Negative, or Zero.

```python
number = int(input('Choose any one number: '))
if(number > 0):
  print('The number chosen is positive')
elif(number < 0):
  print('The number chosen is negative')
else:
  print('The number is zero')
```

Exercise 2: Even or Odd

Ask for a number and say whether it is Even or Odd.

(Hint: Use % operator → number % 2 == 0)

```python
number = int(input('Choose any one number: '))
if(number % 2 == 0):
  print('The number is even')
else:
  print('The number is odd')
```

Today's Mini-Project: "Age & Mood Advisor"

Create a program that:

Asks for the user’s name and age (as number).

Then gives different messages based on age groups:

If age < 13 → "You are a kid. Enjoy your childhood!"

If 13 <= age < 18 → "You are a teenager. Study well but have fun too!"

If 18 <= age < 30 → "You are a young adult. Time to build your future!"

If 30 <= age < 60 → "You are in your prime years!"

If age >= 60 → "Respect! You have a lot of experience."

Add one more question: "How are you feeling today? (happy/sad/tired/excited)"

Then give a short encouraging reply based on their mood (use if-elif).

Use f-strings and make the output look friendly with some lines or emojis.

```python
name = str(input("Hello, what's your name? "))
age = int(input('And your age is? '))

if(age < 13):
  print(f'You are a kid! Enjoy your childhood!')
elif(age >= 13 and age < 18):
  print(f'You are a teenager. Study well but have fun too!')
elif(age >= 18 and age < 30):
  print(f'You are a young adult. Time to build your future!')
elif(age >= 30 and age < 60):
  print(f'You are in your prime years!')
else:
  print(f'Respect! You have a lot of experience.')

mood = str(input('How are you feeling today? (happy/sad/tired/excited) '))

if(mood == 'happy'):
  print(f'You are having a good day!😊')
elif(mood == 'sad'):
  print(f'You will soon find a reason to celebrate ❤️')
elif(mood == 'tired'):
  print(f'You need to rest 😴')
elif(mood == 'excited'):
  print(f'Yeah!! I am celebrating my biggest win today! 🎆')
```
