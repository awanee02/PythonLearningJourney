Dictionary and Set

```python
#Creating a dictionary
student = {"Name" : "Awanee",
           "Age" : 25,
           "City" : "Pune",
           "is_student" : True,
           "marks" : [87,98,94]}

print(student)
print(type(student))

#Accessing values
print(student["Name"])
print(student["City"])
print(student["marks"][1])

#Using .get() method
print(student.get("Name"))
print(student.get("Phone"))   #do not crash if value not present
print(student.get("Phone" "Not found."))

#Adding/updating values
student["Phone"] = '9876543210'
student['Age'] = '22'
print(student)

#Removing items
student.pop("is_student")
del student["Age"]

#Useful dictionary methods
print(student.keys())        #All keys
print(student.values())      #All values
print(student.items())       #All key-value pairs

#Looping through dictionary
for key, value in student.items():
  print(f"{key} : {value}")
```

```python
#Sets

fruits = {"Apple" , "Banana", "Mango", "Guava", "Litchi", "Pineapple"}
print(fruits)

#Useful operations
a = {1,2,3,4}
b = {3,4,5,6}

print(a | b)   #Union
print(a & b)   #Intersection
print(a - b)   #Difference
```

Exercise 3: Word Frequency Counter

Take this sentence:
"python is awesome and python is easy to learn"

Split it into words

Use a dictionary to count how many times each word appears

Print the result

```python
sentence = "python is awesome and python is easy to learn"

#Splitting into words
words = sentence.split()

#Creating an empty dictionary to store the words
word_count = {}

#Actual counting of frequency
for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1

#Printing the result
print("Word frequency: ")
for word, count in word_count.items():
  print(f"{word} : {word_count}")
```

Today's Mini-Project: Personal Profile + Contact Book
Create a program that does the following:

Start with your own profile as a dictionary
Give the user this menu (using while True):

View my profile, Add a new skill, Add a new hobby, Update age, Add a friend to contact book (use another dictionary), View all friends, Exit

Implement all options properly.

```python
# Day 6 Mini-Project - Final Fixed Version

profile = {"name": "Arnav",
           "age": 25,
           "city": "Pune",
          "hobbies": ["coding", "reading", "gaming"],
          "skills": ["Python", "Problem Solving"]}

friends = {}

while True:
    print("PERSONAL PROFILE & CONTACT BOOK")
    print("1. View my profile")
    print("2. Add a new skill")
    print("3. Add a new hobby")
    print("4. Update age")
    print("5. Add a friend to contact book")
    print("6. View all friends")
    print("7. Exit")


    choice = input("\nChoose an option (1-7): ").strip()

    if choice == "1":
        print("\n--- My Profile ---")
        for key, value in profile.items():
            print(f"{key.capitalize()}: {value}")

    elif choice == "2":
        new_skill = input("Enter new skill: ").strip()
        if new_skill:
            profile["skills"].append(new_skill)
            print(f"Skill '{new_skill}' added successfully!")
        else:
            print("Skill cannot be empty!")

    elif choice == "3":
        new_hobby = input("Enter new hobby: ").strip()
        if new_hobby:
            profile["hobbies"].append(new_hobby)
            print(f"Hobby '{new_hobby}' added successfully!")
        else:
            print("Hobby cannot be empty!")

    elif choice == "4":
        try:
            new_age = int(input("Enter new age: "))
            profile["age"] = new_age
            print(f"Age updated to {new_age}")
        except ValueError:
            print("Please enter a valid number for age!")

    elif choice == "5":
        friend_name = input("Enter friend's name: ").strip()
        friend_phone = input("Enter friend's phone number: ").strip()
        if friend_name and friend_phone:
            friends[friend_name] = friend_phone
            print(f"Friend '{friend_name}' added!")
        else:
            print("Name and phone cannot be empty!")

    elif choice == "6":
        if not friends:
            print("You have no friends in the contact book yet.")
        else:
            print("\n--- My Friends ---")
            for name, phone in friends.items():
                print(f"{name} : {phone}")

    elif choice == "7":
        print("Goodbye! Have a great day!")
        break

    else:
        print("Invalid choice! Please enter a number between 1 and 7.")
```
