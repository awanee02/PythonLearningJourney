Today's Mini-Project: Personal Utility Package

Create a small package (folder) with 2 modules:

calculator.py → add, subtract, multiply, divide, power

string_utils.py → capitalize_all, reverse_string, count_words

Then in your main file, import both modules and make a small menu that uses functions from both.

```python
with open("calculator.py","w")as file:
  file.write("""
def add(a,b):
  return a + b

def subtract(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  return a / b

def power(a,b):
  return a ** b
  """)
```

```python
with open("string_utils.py","w")as file:
  file.write("""def capitalize_all(text):
  return text.upper()

def reverse_string(text):
  return text[::-1]

def count_words(text):
  return len(text.split())
  """)

print("Both modules created successfully!")
```

```python
import calculator
import string_utils

print("Add: ", calculator.add(10,20))
print("Subtract: ", calculator.subtract(10,20))
print("Multiply: ", calculator.multiply(10,20))
print("Divide: ", calculator.divide(10,20))
print("Power: ", calculator.power(10,2))

print("Capitalize: ", string_utils.capitalize_all("awanee"))
print("Reverse: ", string_utils.reverse_string("hello"))
print("Word count: ", string_utils.count_words("Python"))
```
