More about functions

```python
x = 10           #Global variable

def my_func():
  y = 20         #Local variable
  print("Inside function: ",x)  #Can access global
  print("Inside function: ",y)

  my_func()
  print("Outside: ",x)
```

```python
#Modifying global variable inside a function
count = 0

def increment():
  global count
  count += 1
  print(count)

increment()
increment()
```

```python
#Default arguments
def greet(name, greeting="Hello!"):
  print(f"{greeting} , {name}")

greet("Arnav")
greet("Awanee", "Good Morning!")
```

*args and **kwargs

```python
#Example with *args
def add_all(*args):
  return sum (args)

print(add_all(1,2,3))
print(add_all(10,20,30,40,50))
```

```python
#Example with **kwargs
def show_info(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} : {value}")

show_info(name = "Awanee", age = 22, city = "Pune", hobby = "Learning new skills")
```

Exercise 1: Scope Practice

Create a global variable total = 0.

Write a function add_to_total(n) that adds n to the global total
using the global keyword.

```python
total = 0

def add_to_total(n):
  global total
  total = total + n
  print(f"Added {n}.Current total is now {total}")
add_to_total(10)
add_to_total(15)
add_to_total(20)
print("Final total: ", total)
```

Exercise 3: *args & **kwargs

Write a function calculate_average(*args) that returns the average of all numbers passed.

Write a function print_details(**kwargs) that prints all key-value pairs passed to it.

```python
def calculate_average(*args):
  if len(args) == 0:
    return 0
  return sum(args)/len(args)
print(calculate_average(1,2,3,4,5))

def print_details(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} : {value}")
print_details(subject = "Python", topic = "functions", mode = "online coding")
```

Today's Mini-Project: Advanced Calculator

Improve your Day 8 calculator with these new features using functions:

Add a function power(base, exponent) for exponentiation (e.g., 2^3 = 8)

Add a function calculate_average(*args) that takes any number of numbers and returns their average

In the menu, add: Power, Average of multiple numbers.

Use default arguments where possible

Make the menu clean and handle invalid inputs gracefully.

```python
def add(a,b):
  return a + b

def subtract(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  if b == 0:
    print("Cannot divide by zero")
  return a / b

def power(base, exponent):
  return base ** exponent

def calculate_average(*args):
  if len(args) == 0:
    return 0
  return sum(args)/len(args)

while True:
  print("1. Addition")
  print("2. Subtraction")
  print("3. Multiplication")
  print("4. Division")
  print("5. Power")
  print("6. Average of multiple numbers")
  print("7. Exit")

  choice = input("Choose an option from (1-7): ")

  if choice == "7":
    print("Options complete")
  break

  if choice == "1":
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = add(num1, num2)
    print(f"{num1} + {num2} = {result}")

  elif choice == "2":
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = subtract(num1, num2)
    print(f"{num1} - {num2} = {result}")


  elif choice == "3":
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = multiply(num1, num2)
    print(f"{num1} * {num2} = {result}")

  elif choice == "4":
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = divide(num1, num2)
    print(f"{num1} / {num2} = {result}")

  elif choice == "5":
    base = float(input("Enter base: "))
    exponent = float(input("Enter exponent: "))
    result = power(base, exponent)
    print(f"{base} ^ {exponent} = {result}")

  elif choice == "6":
    print("Enter numbers separated by space (e.g. 10 20 30 40):")
    numbers = list(map(float, input().split()))
    result = calculate_average(*numbers)   # Unpack list using *
    print(f"Average = {result:.2f}")

  else:
    print("Invalid choice! Please enter a number between 1 and 7.")



```
