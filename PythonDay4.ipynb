for loop with break and continue

```python
for num in range (1,11):
  if num == 5:
    continue          #skip the number and continue with the next iteration
  if num == 8:
    break             #break the loop immediately
  print(num)
```

Exercise 2: Sum of First N Numbers

Ask the user: “How many numbers do you want to add?”

Then ask that many numbers one by one and print their total sum.

```python
#Ask the user how many numbers he wants to add
n = int(input('How many numbers you wwant to add? '))

total = 0  #The total sum after adding each subsequent number
count = 0  #Count of how many numbers we have added

while(count < n):
  number = int(input(f'Enter a number: '))

  total = total + number
  count = count + 1
print(f'The total sum of n numbers is: ', total)
```

Today's Mini-Project: Number Guessing Game

Detailed Requirements:

At the top, write: import random

Computer secretly picks a number: secret = random.randint(1, 50)

Player gets maximum 7 attempts

In each attempt:
Ask user to guess tell them:

“Too high!” if guess > secret

“Too low!” if guess < secret

“Correct!” if guess == secret

Show how many attempts are remaining

If player guesses correctly → Congratulate + tell in how many attempts

If 7 attempts finish without guessing → Reveal the secret number

```python
import random
secret = random.randint(1,50)

attempts_left = 7
guessed_correctly = False

print(f"Welcome to the number guessing game! Let's decide if you are a noob or you are a pro😉")
print(f"I have chosen a number between 1 and 50. You have 7 attempts. Lessgooooo!! \n")

while attempts_left > 0 and not guessed_correctly:
  guess = int(input('Enter your guess: '))

  if  guess == secret:
    attempts_used = 8 - attempts_left
    print(f"You guessed it right! You are definitely a pro ✌️😎")
    print(f"You won in {attempts_used} attempts!")
    guessed_correctly = True
  elif guess > secret:
    print(f"Too high, you are a noob 🤦‍♀️")
  else:
    print(f"Too low, you are a noob 🤦‍♀️")

  attempts_left -= 1

  if attempts_left > 0 and not guessed_correctly:
    print(f"{attempts_left} attempts remaining.")
if not guessed_correctly:
    print(f"No attempts remaining, the secret number is: {secret}")
```
